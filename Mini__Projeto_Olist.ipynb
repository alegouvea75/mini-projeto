```python
# importando as ferramentas que o projeto pediu, biblioteca nativa
import csv
import re
from datetime import datetime

# ler arquivo
def ler_csv(caminho_arquivo):
    dados = []
    with open(caminho_arquivo, encoding='utf-8') as arquivo:
      leitor = csv.DictReader(arquivo)

      for linha in leitor:
        dados.append(linha)
    return dados

products=ler_csv('olist_products_dataset.csv')
orders=ler_csv('olist_orders_dataset.csv')

# carregar o total dos arquivos
print(f"total de products: {len(products)}")
print(f"total de orders: {len(orders)}")
```

```python
# fazer a limpeza dos dados nulos e faltosos "sem categoriads", e 0. Para não descartar as linhas e manter o maior numeros de dados. Pois o projetone pequeno.
def tratar_produtos(dados):

  for item in dados:
    nulos_corrigidos = 0
    if not item["product_category_name"]:
      item["product_category_name"] = "sem categoria"
      nulos_corrigidos += 1

    if not item["product_category_name"]:
      item["product_category_name"] = "sem categoria"

      campos_dimensoes = ["product_weight_g", "product_length_cm", "product_height_cm", "product_width_cm"]
      for campo in campos_dimensoes:
        if not item[campo]:
          item[campo] = 0
  return dados, nulos_corrigidos
```

```python
# Padronização ajustar as letras e espaços.
def padronizar_categorias(dados):
    for item in dados:
      if item["product_category_name"]:
        pcn = item["product_category_name"].lower().strip()
        pcn = re.sub(r'[^\w\s]', '', pcn)
        item["product_category_name"] = pcn
    return dados
```

```python
import csv
import re
from datetime import datetime

# contar os dados cancelados e não entreguados
def validar_pedidos(dados):
  cancelados = 0
  for pedidos in dados:
     if not pedidos["order_delivered_customer_date"] and pedidos["order_status"] == "canceled":
            cancelados += 1
# Arrumado a hora e data para padrão brasileiro
  if pedidos["order_approved_at"]:
      try:
            dt = datetime.strptime(pedidos["order_approved_at"], "%Y-%m-%d %H:%M:%S")
            pedidos["order_approved_at"] = dt.strftime("%d/%m/%Y")
# caso der algum erro continuar
      except ValueError:
          pass

  return dados, cancelados
```

```python
# Relatório final com os parametros que foi pedido.
def relatorio(produtos, pedidos, nulos_corrigidos, cancelados):

  # somar as linhas
  total_linhas = len(produtos) + len(pedidos)
  total_produtos = len(produtos)
  total_pedidos = len(pedidos)

  print("=" * 55)
  print("------------ RELATÓRIO DE STATUS MANUAL ------------")
  print("=" * 55)
  print(f"-> Contagem total de linhas processadas: {total_linhas}")
  print(f"-> Detalhe: {total_produtos} produtos e {total_pedidos} pedidos")
  print(f"-> Total de registros nulos corrigidos: {nulos_corrigidos}")
  print(f"-> Total de pedidos cancelados identificados: {cancelados}")
  print("-" * 55)
  print("STATUS: Base de dados")
  print("=" * 55)


```

```python
# Execução final
print("Iniciando o processamento dos dados")

# Carrehar os dados
dados_produtos = ler_csv('olist_products_dataset.csv')
dados_pedidos = ler_csv('olist_orders_dataset.csv')

# Tratar os produtos
produtos_sem_nulos, total_nulos_corrigidos = tratar_produtos(dados_produtos)
produtos_finais = padronizar_categorias(produtos_sem_nulos)
pedidos_finais, total_pedidos_cancelados = validar_pedidos(dados_pedidos)

# Gerar ralatório
relatorio(produtos_finais, pedidos_finais, total_nulos_corrigidos, total_pedidos_cancelados)

```
